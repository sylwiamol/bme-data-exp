```python
# -*- coding: utf-8 -*-
import glob
import ntpath
import os
import re
import sys

import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile
import scipy.signal
from PyQt5.QtWidgets import *
from scipy.fftpack import fft


# Returns python list of lists in format [beginning_time(int), end_time(int), word(string)]
def parse_timestamps(filename):
	timestamps_file = open(filename)
	list_of_words = timestamps_file.readlines()
	timestamps = list()
	for i in list_of_words:
		t = re.split('\s+', i)
		timestamps.append(
			[float(t[0].replace(',', '.')), float(t[1].replace(',', '.')),
			 t[2]])
	timestamps_file.close()
	return timestamps


# returns signal (THE LEFT SIGNAL CHANNEL IF STEREO) and sampling frequency
def get_signal_from_file(filename):
	fs, signal = scipy.io.wavfile.read(filename)
	if signal.ndim > 1:
		signal = signal[:, 0]
	return signal, fs


if __name__ == '__main__':
	app = QApplication(sys.argv)
	dirname = QFileDialog.getExistingDirectory()
	waves = glob.glob(os.path.join(dirname, '*.wav'))

	first_peaks = []
	words = []
	for wav in waves:
		file_name = os.path.splitext(wav)[0]
		file_base = ntpath.basename(file_name)
		print(file_base)

		signal, fs = get_signal_from_file(file_name + '.wav')
		timestamps = parse_timestamps(file_name + '.txt')
		if not first_peaks:
			# pierwsze uzycie first_peaks
			first_peaks = [[] for x in range(len(timestamps))]

		if not words:
			# pierwsze uzycie words
			words = [x[2] for x in timestamps]

		plt.figure(1)
		plt.title('Signal Wave')
      

		plt.plot(signal)
		# print('Signal length:', len(signal))
		plt.show()

		for i in range(len(timestamps)):
			beginning = int(fs * timestamps[i][0])
			ending = int(fs * timestamps[i][1])

			single_word = signal[beginning:ending]
			# plt.figure(i)

			# plt.title('Single word' + str(i))
			plt.title(file_base+timestamps[i][2])

			len_signal = len(single_word)
			T = 1.0 / 44000.0
			x_signal = np.linspace(0.0, T * len_signal, len_signal)  ########
			# plt.xlim([0, 1200])
			fft_signal = fft(single_word)
			xfft = np.linspace(0.0, 1.0 / (2.0 * T), len_signal // 2)
			yfft = 2.0 / len_signal * np.abs(fft_signal[0:len_signal // 2])
			# plt.plot(xfft, yfft)

			# zapisywanie pliku z wykresem
			# plt.savefig(os.path.join(dirname, 'Word_' + str(file_base) + '_' + str(i)))
			# print(len_signal)
			# print(max(fft_signal))
			# print(len(yfft))
			zakres = 100
			peaks = []
			for j in range(zakres + 1, len(yfft) - zakres - 1):
				if max(yfft[j - zakres:j + zakres]) == yfft[j]:
					# plt.plot(j, yfft[j], marker="o", markersize=7)
					# print('PIK: ' + str(yfft[j]))
					peaks.append(yfft[j])
			if peaks:
				first_peaks[i].append(peaks[0])
		len_signal = len(signal)
		T = 1.0 / 44000.0
		x_signal = np.linspace(0.0, T * len_signal, len_signal)

		# plt.figure(3)

		fft_signal = fft(signal)
		xfft = np.linspace(0.0, 1.0 / (2.0 * T), len_signal // 2)

		y = 2.0 / len_signal * np.abs(fft_signal[0:len_signal // 2])
		# plt.plot(xfft, y)
		# print("LEN FFT2", len(fft_signal))

		# plt.show()

	s_info = []
	nowe=37.6176161558
	slowo=words[0]
	roznica=1000
	for i in range(len(first_peaks)):
		# srednia z pierwszego piku dla kazdego slowa
		s = "%s: %s" % (words[i], sum(first_peaks[i]) / len(first_peaks[i]))
		if abs(sum(first_peaks[i]) / len(first_peaks[i])-nowe)<roznica:
			slowo=words[i]
			roznica=abs(sum(first_peaks[i]) / len(first_peaks[i])-nowe)
		s_info.append(s)
	s_info = '\n'.join(s_info)
	print(slowo)
	#QMessageBox.information(None, "Wynik", s_info)

	sys.exit(app.exec_())
```
