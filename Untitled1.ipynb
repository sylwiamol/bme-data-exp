```python
import pandas as pd
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import csv
import os
import glob
import itertools
## data = wave.open('266725_22_K_23_4.wav', 'r')
#signal = data.readframes(-1)
#signal = np.fromstring(signal, 'Int16')
#plt.figure(1)
#plt.title('Signal Wave...')
#plt.plot(signal)
#plt.show()

#text_file = open("266725_22_K_23_4.txt", "r")
#print(text_file.read())

#lines = text_file.read()
#lines2=re.split('\t|\n', lines)
#print (lines2)


#na podstawie: https://www.e-education.psu.edu/geog485/node/141
dane = glob.glob("/home/sylwia/bme-data-exp/Pliki/*.wav")
opisy = glob.glob("/home/sylwia/bme-data-exp/Pliki/*.txt")
data_list = list()
command_list = list()

for x in range (0, len(dane)):
    sample, data = wav.read(dane[x], mmap=False)
    labels = list(csv.reader(open(opisy[x], 'r'), delimiter='\t'))
 # dzwiek w niektorych plikach jest stereo wiec uwzgledniamy tylko jeden kanal  wiec zero na koncu, mozna tez srednia z dwoch kanalow   
    for i in range (0,len(labels)):
        data_list.append(data[int(float(labels[i][0])*sample)-1: int(float(labels[i][1])*sample)+1,0])
        command_list.append(labels[i][2])



d = {'komenda': command_list, 'dane': data_list}

#constructing dataframe: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.html
df = pd.DataFrame.from_records(data=d)
print(df['komenda'].value_counts())

df

```
