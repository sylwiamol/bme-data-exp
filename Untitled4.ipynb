```python

import glob
import ntpath
import os
import re
import sys

import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile
import scipy.signal
from PyQt5.QtWidgets import *
from scipy.fftpack import fft


# Returns python list of lists in format [beginning_time(int), end_time(int), word(string)]
def parse_timestamps(filename):
	timestamps_file = open(filename)
	list_of_words = timestamps_file.readlines()
	timestamps = list()
	for i in list_of_words:
		t = re.split('\s+', i)
		timestamps.append(
			[float(t[0].replace(',', '.')), float(t[1].replace(',', '.')),
			 t[2]])
	timestamps_file.close()
	return timestamps


# returns signal (THE LEFT SIGNAL CHANNEL IF STEREO) and sampling frequency
def get_signal_from_file(filename):
	fs, signal = scipy.io.wavfile.read(filename)
	if signal.ndim > 1:
		signal = signal[:, 0]
	return signal, fs


if __name__ == '__main__':
	app = QApplication(sys.argv)
	dirname = QFileDialog.getExistingDirectory()
	waves = glob.glob(os.path.join(dirname, '*.wav'))
	print(len(waves))
	srednie = []
	words = []
	odch = []
	odl=[]
	for wav in waves:
		file_name = os.path.splitext(wav)[0]
		file_base = ntpath.basename(file_name)
		print(file_base)
		signal, fs = get_signal_from_file(file_name + '.wav')
		timestamps = parse_timestamps(file_name + '.txt')
#normalizacja sygnalow
		maksimum=np.amax(signal)
		signal=signal*(1/maksimum)
		if not srednie:
			srednie = [0 for x in range(len(timestamps))]
		if not odch:
			odch = [0 for x in range(len(timestamps))]
		if not words:
			words = [x[2] for x in timestamps]
		if not odl:
			odl = [0 for x in range(len(timestamps))]
#plotowanie wszystkich sygnalow
		plt.figure(1)
		plt.title('Signal Wave')
      

		plt.plot(signal)

#		plt.show()
		srednia=0
		odchylenie=0
		odleglosc=0
		for i in range(len(timestamps)):
			beginning = int(fs * timestamps[i][0])
			ending = int(fs * timestamps[i][1])

			single_word = signal[beginning:ending]
			srednia=single_word.mean()
			odchylenie=single_word.std()
			odleglosc=single_word.ptp()
			srednie[i]=(srednie[i]+srednia)
			odch[i]=(odch[i]+odchylenie)
			odl[i]=(odl[i]+odleglosc)

	s_info = []


	for i in range(len(timestamps)):
		# srednia z srednich, odchylenia standardowego i odleglosci max-min z kazdego slowa
		srednie[i]=srednie[i]/len(waves)
		odch[i]=odch[i]/len(waves)
		odl[i]=odl[i]/len(waves)
		s = "%s: %s %s %s" % (words[i], srednie[i], odch[i], odl[i])
# 		if abs(sum(first_peaks[i]) / len(first_peaks[i])-nowe)<roznica:
#			slowo=words[i]
#			roznica=abs(sum(first_peaks[i]) / len(first_peaks[i])-nowe)
		s_info.append(s)
	s_info = '\n'.join(s_info)
	print(s_info)

    
    
    
    
## sygnal testowy
    
	signal2, fs2 = get_signal_from_file('266725_22_K_23_4.wav')
	maksimum2=np.amax(signal2)
	signal2=signal2*(1/maksimum2)

#Parse Timestamps file into a proper list
	timestamps2 = parse_timestamps('266725_22_K_23_4.txt')



	words2=[]
	s_info2 = []
	if not words2:
	# pierwsze uzycie words
		words2 = [x[2] for x in timestamps2]
	for i in range(len(timestamps2)):
		beginning2 = int(fs2*timestamps2[i][0])
		ending2 = int(fs2*timestamps2[i][1])

		single_word2 = signal2[beginning2:ending2]
		#env = get_envelope(single_word, 250)
		s2="%s: %s %s %s" % (words2[i],  single_word2.mean(), single_word2.std(), single_word2.ptp())
		s_info2.append(s2)

	s_info2 = '\n'.join(s_info2)
	print('dla sygnalu testowego')
	print(s_info2)
#	plt.show()

	sys.exit(app.exec_())
```
