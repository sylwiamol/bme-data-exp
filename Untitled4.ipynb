```python

import glob
import ntpath
import os
import re
import sys
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile
import scipy.signal
from PyQt5.QtWidgets import *
from scipy.fftpack import fft


##  [rozp(int), zakonczenie(int), slowo(string)]
def labele(plik):
	labelki_file = open(plik)
	list_of_words = labelki_file.readlines()
	labelki = list()
	for i in list_of_words:
		t = re.split('\s+', i)
		labelki.append(
			[float(t[0].replace(',', '.')), float(t[1].replace(',', '.')),
			 t[2]])
	labelki_file.close()
	return labelki


# jeden sygnal jesli dzwiek stereo (np moje pliki)
def sygnal(plik):
	fs, signal = scipy.io.wavfile.read(plik)
	if signal.ndim > 1:
		signal = signal[:, 0]
	return signal, fs


if __name__ == '__main__':
# otwarcie pliku, done by Marta M. 
	app = QApplication(sys.argv)
	dirname = QFileDialog.getExistingDirectory()
	waves = glob.glob(os.path.join(dirname, '*.wav'))
#inicjalizacja list do parametrow
	srednie = []
	words = []
	odch = []
	odl=[]
#wszystkie pliki po kolei liczymy
	for wav in waves:
		file_name = os.path.splitext(wav)[0]
		file_base = ntpath.basename(file_name)
		signal, fs = sygnal(file_name + '.wav')
		labelki =labele(file_name + '.txt')
#normalizacja sygnalow
		maksimum=np.amax(signal)
		signal=signal*(1/maksimum)
#inicjalizacja list dla poszczegolnych wyliczanych parametrow, uzupelnienie listy slow
		if not srednie:
			srednie = [0 for x in range(len(labelki))]
		if not odch:
			odch = [0 for x in range(len(labelki))]
		if not words:
			words = [x[2] for x in labelki]
		if not odl:
			odl = [0 for x in range(len(labelki))]
#plotowanie wszystkich sygnalow
		plt.figure(1)
		plt.title('Signal Wave')
		plt.plot(signal)
#		plt.show()
#zmienne pomocniczne
		srednia=0
		odchylenie=0
		odleglosc=0
#podzial na slowa i obliczanie parametrow
		for i in range(len(labelki)):
			poczatek = int(fs * labelki[i][0])
			koniec = int(fs * labelki[i][1])
# dodanie do sumarycznej wartosci, wartosci z kolejnych nagran
			single_word = signal[poczatek:koniec]
			srednia=single_word.mean()
			odchylenie=single_word.std()
			odleglosc=single_word.ptp()
			srednie[i]=(srednie[i]+srednia)
			odch[i]=(odch[i]+odchylenie)
			odl[i]=(odl[i]+odleglosc)

	s_info = []
	for i in range(len(labelki)):
		# srednia z srednich, odchylenia standardowego i odleglosci max-min z kazdego slowa
		srednie[i]=srednie[i]/len(waves)
		odch[i]=odch[i]/len(waves)
		odl[i]=odl[i]/len(waves)
		s = "%s: %s, %s, %s" % (words[i], srednie[i], odch[i], odl[i])
# 		if abs(sum(first_peaks[i]) / len(first_peaks[i])-nowe)<roznica:
#			slowo=words[i]
#			roznica=abs(sum(first_peaks[i]) / len(first_peaks[i])-nowe)
		s_info.append(s)
	s_info = '\n'.join(s_info)
	print('Srednie wartosci poszczegolnych parametrow ze wszystkich sygnalow uczacych')  
	print('SLOWO: srednia, odchylenie standardowe, odleglosc peak-to-peak')
	print(s_info)

    
    
    
    
## sygnal testowy
    
	signal2, fs2 = sygnal('266725_22_K_23_4.wav')
	maksimum2=np.amax(signal2)
	signal2=signal2*(1/maksimum2)
	labelki2 =labele('266725_22_K_23_4.txt')
	words2=[]
	s_info2 = []
	if not words2:
		words2 = [x[2] for x in labelki2]
	for i in range(len(labelki2)):
		poczatek2 = int(fs2*labelki2[i][0])
		koniec2 = int(fs2*labelki2[i][1])

		single_word2 = signal2[poczatek2:koniec2]
		s2="%s: %s %s %s" % (words2[i],  single_word2.mean(), single_word2.std(), single_word2.ptp())
		s_info2.append(s2)

	s_info2 = '\n'.join(s_info2)
	print('Srednie wartosci poszczegolnych parametrow dla wszystkich sygnalow testowych')  
	print('SLOWO: srednia, odchylenie standardowe, odleglosc peak-to-peak')
	print(s_info2)

	sys.exit(app.exec_())
```

- odchylenie standardowe
- odległość maksymalnej i minimalnej wartości
